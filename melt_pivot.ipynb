```python
import pandas as pd
```

```python
df = pd.read_csv('car.csv')
```

```python
df
```

```python
df[['fueltype','aspiration','peakrpm']]
```

```python
df.pivot_table(index='fueltype',columns='aspiration',values='peakrpm',aggfunc='sum')
```

```python
df[['fueltype','aspiration','peakrpm']].melt()
```

```python
df.melt(id_vars='fueltype',value_vars='aspiration')
```
